```sql
CREATE PROCEDURE show_country_results AS
SELECT * FROM test3.dbo.new_table
```

```sql
ALTER PROCEDURE show_country_results @country NVARCHAR(30), @year SMALLINT, @gender NVARCHAR(50), @medaltype NVARCHAR(50)
AS
SELECT * FROM test3.dbo.new_table
WHERE NOC = @country
and Edition = @year
and Gender = @gender
and Medal = @medaltype
```
