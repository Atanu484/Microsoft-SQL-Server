```sql
-- This query shows how many countries got how many gold medels in which decades and in which event and in which gender
SELECT
decades,
NOC,
Gender,
Event,
COUNT(MedalistID) as gold_medalists
FROM
(
SELECT 
Edition / 10 * 10 as decades,
*
FROM test3.dbo.new_table
WHERE MedalValue IN
 (SELECT MedalValue
  FROM test3.dbo.new_table
  WHERE MedalValue = 1)
)a
GROUP BY NOC, decades, Gender, Event
ORDER BY decades
```
