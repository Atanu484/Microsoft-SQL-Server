```sql
-- Ranks the countries by their sportsmenship
WITH 
finals AS (
    SELECT
    NOC,
    COUNT(DISTINCT Edition) AS number_of_final_qualifications
   FROM test3.dbo.new_table
   GROUP BY NOC)
, gold AS (
    SELECT
    NOC,
    COUNT(MedalistID) AS gold_medalistID
    FROM test3.dbo.new_table
    WHERE Medal = 'Gold'
   GROUP BY NOC
),
silver AS (
    SELECT
    NOC,
    COUNT(MedalistID) AS silver_medalistID
    FROM test3.dbo.new_table
    WHERE Medal = 'Silver'
   GROUP BY NOC
),
bronze AS (
    SELECT
    NOC,
    COUNT(MedalistID) AS bronze_medalistID
    FROM test3.dbo.new_table
    WHERE Medal = 'Bronze'
   GROUP BY NOC
)

SELECT
a.NOC,
number_of_final_qualifications,
b.gold_medalistID * 3  AS Gold,
c.silver_medalistID * 2 AS Silver,
d.bronze_medalistID AS Bronze, -- Multipying with 1 will lead to the current result
gold_medalistID * 3 + silver_medalistID * 2+ bronze_medalistID AS Competetiveness,
(gold_medalistID * 3 + silver_medalistID * 2 + bronze_medalistID) / number_of_final_qualifications AS Ranking
FROM finals a
JOIN gold b
ON a.NOC = b.NOC
JOIN silver c 
ON a.NOC = c.NOC
JOIN bronze d 
ON a.NOC = d.NOC
ORDER BY Competetiveness DESC
```
